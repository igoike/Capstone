```python
import numpy as np
import pandas as pd
import re
import os 
import glob
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
```

```python
data_file_folder ='../data/'
df = []
for file in os.listdir(data_file_folder):
        if file.endswith('.xlsx'):             
             df.append(pd.read_excel(os.path.join(data_file_folder,file),sheet_name='Sheet1'))
df_master = pd.concat(df, axis = 0)
```

```python
df_master.isna().sum()
```

```python
df_master
```

```python
df_master.to_excel('../data/master_file.xlsx', engine='xlsxwriter',index=False)
```

```python
df_master['injuries'] = pd.to_numeric(df_master['injuries'],errors='coerce')
df_master['total_crashes'] = pd.to_numeric(df_master['total_crashes'],errors='coerce')
df_master
```

```python
df_master.info()
```

```python
df_master_pop_gr = df_master.groupby(["CBSA Title","Year","Metropolitan/Micropolitan Statistical Area"])["population"].sum().reset_index()
df_master_pop_gr
```

```python
df_master_pop_gr = pd.pivot(df_master_pop_gr,index=["CBSA Title","Metropolitan/Micropolitan Statistical Area"],columns=["Year"])
```

```python
df_master_pop_analysis = df_master_pop_gr.pct_change(axis='columns')
df_master_pop_analysis
```

```python
df_master_fatal_gr = df_master.groupby(["CBSA Title","Year","Metropolitan/Micropolitan Statistical Area","State","population"])["fatalities"].sum().reset_index()
df_master_fatal_gr
```

```python
df_master_total_gr = df_master.groupby(["CBSA Title","Year","Metropolitan/Micropolitan Statistical Area","State","population"])["total_crashes"].sum().reset_index()
df_master_total_gr 
```

```python
df_master_inj_gr = df_master.groupby(["CBSA Title","Year","Metropolitan/Micropolitan Statistical Area","State","population"])["injuries"].sum().reset_index()
df_master_inj_gr 
```

```python
df_master_fatal_gr['Fatalities per 100,000 population'] = round((df_master_fatal_gr['fatalities'] / df_master_fatal_gr['population']) * 100000)
df_master_inj_gr['Injuries per 100,000 population'] = round((df_master_inj_gr['injuries'] / df_master_inj_gr['population']) * 100000)
df_master_total_gr['Total Crashes per 100,000 population'] = round((df_master_total_gr['total_crashes'] /df_master_total_gr['population']) * 100000)
```
